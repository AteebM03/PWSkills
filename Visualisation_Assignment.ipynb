```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
# variable across different categories.

np.random.seed(42)
categories = np.random.choice(['Category A', 'Category B', 'Category C'], size=300)
values = np.random.normal(loc=0, scale=1, size=300) + (categories == 'Category B') * 1.5 + (categories == 'Category C') * 3

df = pd.DataFrame({'Category': categories, 'Value': values})

sns.boxplot(data=df, x='Category', y='Value', palette='Set2')

plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.

np.random.seed(42)
n = 100
X1 = np.random.randn(n)
X2 = X1 * 0.8 + np.random.randn(n) * 0.2
X3 = X1 * 0.3 + np.random.randn(n) * 0.7
X4 = np.random.randn(n)

df = pd.DataFrame({'X1': X1, 'X2': X2, 'X3': X3, 'X4': X4})

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# PLOTLY ASSIGNMENT:
# 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space.

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

import plotly.express as px

fig = px.scatter_3d(df, x='X', y='Y', z='Z', title="3D Scatter Plot of Data Points",
                    labels={'X': 'X Axis', 'Y': 'Y Axis', 'Z': 'Z Axis'})
fig.show()
```

```python
# 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade
# categories.

np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

fig = px.violin(df, x='Grade', y='Score', box=True, points="all",
                title="Distribution of Scores Across Different Grade Categories",
                labels={'Grade': 'Grade Category', 'Score': 'Score'},
                color='Grade',
                category_orders={'Grade': ['A', 'B', 'C', 'D', 'F']})

fig.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
# days.

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

pivot_df = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='sum')

fig = px.imshow(pivot_df,
                labels={'x': 'Month', 'y': 'Day', 'color': 'Sales'},
                title='Sales Heatmap Across Different Months and Days',
                color_continuous_scale='Viridis')
fig.show()
```

```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(** 0.5 x**2 + y**2)

import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(
    title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))",
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    autosize=True
)
fig.show()
```

```python
# BOKEH ASSIGNMENT:
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', width=600, height=400)

p.line(x, y, legend_label="Sine Wave", line_width=2, line_color="blue")

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

from bokeh.models import ColumnDataSource

output_notebook()

np.random.seed(42)
n = 100
x = np.random.random(n) * 10
y = np.random.random(n) * 10
sizes = np.random.randint(5, 50, size=n)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)

source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=600, height=400)

p.circle('x', 'y', size='sizes', color='colors', legend_field='colors', source=source, alpha=0.6)

p.legend.title = 'Color Legend'
p.legend.label_text_font_size = '10pt'
p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_notebook()

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

source = ColumnDataSource(data={'fruits': fruits, 'counts': counts})

p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count',
           width=600, height=400)
p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field="fruits", color="green", alpha=0.7)

p.xaxis.major_label_orientation = "vertical"
p.legend.title = 'Fruit Type'
p.legend.label_text_font_size = '10pt'
p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

output_notebook()
source = ColumnDataSource(data={'top': hist, 'left': edges[:-1], 'right': edges[1:], 'bottom': np.zeros_like(hist)})

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=600, height=400)

p.quad(top='top', bottom='bottom', left='left', right='right', source=source, color="skyblue", alpha=0.7)

p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import LinearColorMapper

output_notebook()

x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=600, height=600)

color_mapper = LinearColorMapper(palette=Viridis256, low=min(z_flat), high=max(z_flat))

p.image(image=[z_flat], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, width=8, location=(0,0))
p.add_layout(color_bar, 'right')

p.xaxis.axis_label_text_font_size = '12pt'
p.yaxis.axis_label_text_font_size = '12pt'

show(p)
```
