```python
#6. Explain the concept of generators in Python and how they are defined.

#Generators: Generators are a type of iterable in Python that allow you to create a sequence of values lazily
# using the yield statement to produce values one at a time and maintain state between iterations.

#Defining Generators: Generators are defined using a function with the yield keyword instead of return
# which allows the function to produce a series of values on demand.

#Example:
def fib(n):
  '''The Fibonacci sequence '''
  a, b = 0, 1
  for i in range(n):
    yield a
    a, b = b, a + b
```

```python
#7. What are the advantages of using generators over regular functions?

#Here are some key advantages of using generators over regular functions:

# **Memory Efficiency**: Generators produce values one at a time and only when needed, which can
# be more memory-efficient than generating an entire sequence at once.

# **Lazy Evaluation**: Generators compute values on-the-fly, allowing for processing of
# large data sets or infinite sequences without the need to store the entire sequence in memory.

# **Improved Performance**: By generating values as needed, generators can reduce the overhead
# associated with creating and managing large collections of data.

# **Simplified Code**: Generators can simplify code that requires iteration over large sequences
# as they eliminate the need for explicit state management and can be written in a more concise form.
```

```python
#10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

#Here's a brief comparison of the map(), reduce(), and filter() functions in Python:

#map():

#Purpose: Applies a function to each item in an iterable and returns a map object with the results.
#Usage: map(function, iterable)
#Example: map(lambda x: x * 2, [1, 2, 3]) produces [2, 4, 6].
list(map(lambda x: x * 2, [1, 2, 3]))
```

```python
#reduce() (from the functools module):

#Purpose: Applies a function cumulatively to the items of an iterable, reducing it to a single value.
#Usage: reduce(function, iterable)
#Example: reduce(lambda x, y: x + y, [1, 2, 3, 4]) produces 10.
from functools import reduce
reduce(lambda x, y: x + y, [1, 2, 3, 4])

```

```python
#filter():

#Purpose: Applies a function to each item in an iterable and returns an iterator with items for which the function returns True.
#Usage: filter(function, iterable)
#Example: filter(lambda x: x % 2 == 0, [1, 2, 3, 4]) produces [2, 4].
list(filter(lambda x: x % 2 == 0, [1, 2, 3, 4]))
```

```python
#Practical Questions:
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even(l):
  """returns sum of all even numbers in a list"""
  return sum(i for i in l if i%2 == 0)
```

```python
#3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def sq_num(nums):
  """returns list containing the squares of each number"""
  l = []
  for i in nums:
    l.append(i**2)
  return l
```

```python
nums = [2, 5, 6, 3, 9]
sq_num(nums)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True  # 2 is the only even prime number
    if n % 2 == 0:
        return False  # All other even numbers are not prime
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

```

```python
is_prime(4)
```

```python
is_prime(29)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temp = [25, 56, 74, 12]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temp))
```

```python
#11) Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:


#Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the
#order is smaller than 100,00 â‚¬.

#Write a Python program using lambda and map.

order_no = [1, 2, 3, 4, 5]
price_per_item = [30, 20, 50, 70, 10]
quantity = [3, 5, 1, 2, 4]
total_price = list(map(lambda price, qty: price * qty if (price * qty) >= 100 else (price * qty) + 10, price_per_item, quantity))
result = []
for i in range(0, len(order_no)):
  result.append((order_no[i], total_price[i]))
print(result)
```
